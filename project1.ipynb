### Correlation:  Plotting the distributions (Total, Higgs Boson and Other particules) for each feature

These plots enable us to get a feeling for the distribution of the features.

```python
def plotHistogram(index, title,labels):
    condition = np.logical_or(np.full(labels.shape, False, dtype=bool), (tx[:,index]>-990))
    fig, axs = plt.subplots(1, 3,figsize=(16,5))
    axs[0].hist(tx[condition,index], color="b")
    axs[0].set_title('Total Distribution')
    axs[1].hist(tx[np.logical_and(labels==-1, condition),index], color='r')
    axs[1].set_title('Higgs Boson')
    axs[2].hist(tx[np.logical_and(labels==1, condition),index], color='g')
    axs[2].set_title('Other particules')

    plt.suptitle(title)
    plt.show()

for index in range(tx.shape[1]):
    plotHistogram(index,names[index],y)
```

## Data preprocessing

From here on, we present the final result of our reasoning and the steps used for our best prediction. Our reasoning is presented in more detail in our report.

### Divinding the data into groups 

In each group, the same variables are defined and remove the undefined variables in each group. This enables us to effectively handle the Nan values (-999). We will be able to use the prediction on any test set using the same groups and we won't need to exclude the Nan values (-999).

```python
from implementations import *
#Reloading the data to reset previous preprocessing
y, x, ids = load_csv_data(DATA_PATH_TRAIN)
tx, ids = divide_in_groups(x)
```

## Add an offeset 
Add a constant feature with value 1 to allow the separating hyperplane to shift from the origin


```python
shape01 = tx[0].shape[1]
shape12 = tx[3].shape[1]

usable_tx01 = add_offset(polynomial_embedding(tx[0][:,0:shape01-2],3))
usable_tx11 = add_offset(polynomial_embedding(tx[1],3))
usable_tx02 = add_offset(polynomial_embedding(tx[2],3))
usable_tx12 = add_offset(polynomial_embedding(tx[3][:,0:shape12-2],3))
usable_tx03 = add_offset(polynomial_embedding(tx[4],3))
usable_tx13 = add_offset(polynomial_embedding(tx[5],3))
```

## Cross validation for the hyperparameters

We used cross validation to find the optimal regularization parameter and degrees for our polynomial embedding

```python
lambda01 = cross_validation_demo(y[ids[0]],usable_tx01)
lambda11 = cross_validation_demo(y[ids[1]],usable_tx11)
lambda02 = cross_validation_demo(y[ids[2]],usable_tx02)
lambda12 = cross_validation_demo(y[ids[3]],usable_tx12)
lambda03 = cross_validation_demo(y[ids[4]],usable_tx03)
lambda13 = cross_validation_demo(y[ids[5]],usable_tx13)

print(lambda01,lambda02,lambda03,lambda11,lambda12,lambda13)

degrees01 = bias_variance_demo(y[ids[0]],tx[0][:,0:shape01-2],lambda01)
degrees11 = bias_variance_demo(y[ids[1]],tx[1],lambda11)
degrees02 = bias_variance_demo(y[ids[2]],tx[2],lambda02)
degrees12 = bias_variance_demo(y[ids[3]],tx[3][:,0:shape12-2],lambda12)
degrees03 = bias_variance_demo(y[ids[4]],tx[4],lambda03)
degrees13 = bias_variance_demo(y[ids[5]],tx[5],lambda13)

print("The best degrees are :",degrees01,degrees11,degrees02,degrees12,degrees13)
```

### Polynomial embedding 

We used the degree that resulted from the relevant cross-validation 

```python
usable_tx01 = add_offset(polynomial_embedding(tx[0][:,0:shape01-2],degrees01))
usable_tx11 = add_offset(polynomial_embedding(tx[1],degrees11))
usable_tx02 = add_offset(polynomial_embedding(tx[2],degrees02))
usable_tx12 = add_offset(polynomial_embedding(tx[3][:,0:shape12-2],degrees12))
usable_tx03 = add_offset(polynomial_embedding(tx[4],degrees03))
usable_tx13 = add_offset(polynomial_embedding(tx[5],degrees13))
```

### Model fitting: Ridge regression

```python
w01,loss01 = ridge_regression(y[ids[0]],usable_tx01, lambda01)
w11,loss11 = ridge_regression(y[ids[1]],usable_tx11, lambda12)
w02,loss02 = ridge_regression(y[ids[2]],usable_tx02,lambda02)
w12,loss12 = ridge_regression(y[ids[3]],usable_tx12,lambda12)
w03,loss03 = ridge_regression(y[ids[4]],usable_tx03,lambda03)
w13,loss13 = ridge_regression(y[ids[5]],usable_tx13,lambda13)
```
